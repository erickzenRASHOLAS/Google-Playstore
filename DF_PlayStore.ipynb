**METEMOS TODO LO QUE LLEVAMOS HASTA EL MOMENTO EN UN PIPELINE PARA NO TENER QUE EJECUTAR TODO EL CODIGO PARTE POR PARTE**

```
import pandas as pd
import numpy as np
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
```

```
df=pd.read_csv("googleplaystore.csv")
df_clean=df.copy()
df_clean.isnull().sum()
```

```
def pipeline_limpieza(df_input):
    df = df_input.copy()

    # 1. Eliminar duplicados y resetear índice
    df = df.drop_duplicates().reset_index(drop=True)

    # 2. Eliminar la app específica por nombre (Fila corrupta)
    # Esto es más seguro que usar .drop(9990)
    df = df[df['App'] != 'Life Made WI-Fi Touchscreen Photo Frame'].reset_index(drop=True)

    # 3. Transformaciones de Outliers en Installs y Android Ver
    df['Installs'] = df['Installs'].str.replace('0+', '0', regex=False)
    df['Android Ver'] = df['Android Ver'].str.replace('4.4W and up', '4.4 and up', regex=False)

    #4.Cambiamos el formato de las fechas a Date
    df['Last Updated'] = pd.to_datetime(df['Last Updated'], errors='coerce')

    # 5. Tratamiento de nulos (Imputación por Moda)
    df['Type'] = df['Type'].fillna(df['Type'].mode()[0])
    df['Current Ver'] = df['Current Ver'].fillna(df['Current Ver'].mode()[0])
    df['Android Ver'] = df['Android Ver'].fillna(df['Android Ver'].mode()[0])

    # 6. Corrección lógica: Rating = 0 cuando no hay instalaciones y es nulo
    condicion_nulos = (df['Rating'].isnull()) & (df['Installs']=="0")
    df.loc[condicion_nulos, 'Rating'] = 0

    #7. Reemplazamos Los Rating Nulos Por la median que es menos sensible a Outliners
    df['Rating'] = df['Rating'].fillna(df['Rating'].median())

    #8.Reemplazamos el precio por Float
    df['Price'] = df['Price'].str.replace('$', '', regex=False).astype(float)

    #9. Transformación con OneHotEncoder (Gratis=0, Paga=1)
    # drop='first' con Free/Paid dejará 'Paid' como la columna activa (1)
    encoder = OneHotEncoder(drop='first', sparse_output=False)
    df['Type'] = encoder.fit_transform(df[['Type']])

    # 10 Limpieza de 'Installs' (Eliminar '+' y ',')
    df['Installs'] = df['Installs'].str.replace('+', '').str.replace(',', '')
    # Manejar posibles errores (como la fila corrupta si es que existe)
    df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce').fillna(0).astype(int)


    # 12 Limpieza de 'Size' (Convertir todo a Megabytes)
    def clean_size(size):
        if 'M' in size:
            return float(size.replace('M', ''))
        elif 'k' in size:
            return float(size.replace('k', '')) / 1024
        elif 'Varies with device' in size:
            return np.nan
        return np.nan

    df['Size'] = df['Size'].apply(clean_size)

    # 13. Aplicar StandardScaler
    scaler = StandardScaler()
    # Creamos una nueva columna para no perder el valor original en MB
    df['Size_Scaled'] = scaler.fit_transform(df[['Size']])

    return df

#Ejecucion
df_clean = pipeline_limpieza(df)

# Verificación
print(f"¿Existe la app corrupta?: { 'Sí' if 'Life Made WI-Fi Touchscreen Photo Frame' in df_clean['App'].values else 'No'}")
print(f"Nulos en Rating después del ajuste: {df_clean['Rating'].isnull().sum()}")
```

```
import pandas as pd
import numpy as np

# 2. Limpieza de 'Size' (Convertir todo a Megabytes)
def clean_size(size):
    if 'M' in size:
        return float(size.replace('M', ''))
    elif 'k' in size:
        return float(size.replace('k', '')) / 1024
    elif 'Varies with device' in size:
        return np.nan
    return np.nan

df['Size'] = df['Size'].apply(clean_size)

# 5. Estandarización de Texto (Categorías y Géneros a minúsculas para evitar duplicados por formato)
df['Category'] = df['Category'].str.lower()
df['Genres'] = df['Genres'].str.lower()


print("Estandarización completada. Tipos de datos finales:")
print(df.dtypes)
df.head()
```
